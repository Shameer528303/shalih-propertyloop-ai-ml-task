```python
import pandas as pd

a = pd.read_csv("/content/trades.csv")
b = pd.read_csv("/content/holdings.csv")

print("Trades:", a.shape)
print("Holdings:", b.shape)

print("Trades columns:", a.columns.tolist())
print("Holdings columns:", b.columns.tolist())

a.columns = a.columns.str.strip()
b.columns = b.columns.str.strip()

def data_extr(c):
    d = c.lower()

    if "holding" in d and "total" in d:
        for i in b["PortfolioName"].astype(str).unique():
            if i.lower() in d:
                h = b[b["PortfolioName"] == i].shape[0]
                return f"{i} has {h} holdings."

    if "trade" in d and "total" in d:
        for i in a["PortfolioName"].astype(str).unique():
            if i.lower() in d:
                g = a[a["PortfolioName"] == i].shape[0]
                return f"{i} has {g} holdings."

    if "performed" in d or "profit" in d or "p&1" in d:
        e = b.groupby("PortfolioName")["PL_YTD"].sum()
        k = e.idxmax()
        return f"{k} performed best with Year-to-Date P&L = {e.max():,.2f}"

    return None

def fn_chatbot(q):
    out = data_extr(q)
    if out is None:
        return "Sorry can not find the answer"
    return out
```

```python
questions = [
    "Total number of trades for Redfield Accu-Fund",
    "Total number of holdings for MNC Investment Fund",
    "Which fund performed better",
    "Who is the father of science?"
]

for i in questions:
    print(f"Q: {i}\nA: {fn_chatbot(i)}\n")
```
